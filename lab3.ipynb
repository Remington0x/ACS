## Модель «хищник – жертва»

```python
var("y1 y2")

t = srange(0, 7, 0.1)
y0 = [2, 1]
a, c, d = 3, 1, 1

plt = line([])
for b in range(4,1,-1):
    eq = [y1 * (a - b * y2), 
          y2 * (-c + d * y1)]
    sols = desolve_odeint(eq, y0, t, dvars=[y1, y2])
    plt += line(sols, rgbcolor=Color(randrange(100)/100, 1, 1, space='hsv').rgb())
    
plt.show(aspect_ratio = 1)
```

## Падение тела

```python
var("y1 y2 y3 y4 t")

g = 9.81
k = 0.01

eq = [y2,
      -k * y2 * sqrt(y2**2 + y4**2),
      y4,
      -k * y4 * sqrt(y2**2 + y4**2) - g]

tmax = 1.41
alph = pi / 4
v0 = 10.0
y0 = [0, v0 * cos(alph), 0, v0 * sin(alph)]
t0 = 0

sols = desolve_system_rk4(eq, [y1, y2, y3, y4], ics=[t0, *y0], ivar=t, step=0.1, end_points=tmax)

plt = line([])
for i in range(len(sols) - 1):
    plt += line([(sols[i][1], sols[i][3]), (sols[i+1][1], sols[i+1][3])])

plt.show()
```
